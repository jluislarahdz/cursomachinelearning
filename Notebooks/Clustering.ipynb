## Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
url_movies = "https://raw.githubusercontent.com/jluislarahdz/cursomachinelearning/luislaramachine/movies/movies.csv"
```

```python
data = pd.read_csv(url_movies, sep =";")
```

```python
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1) # m√©trica manhattan
dd2 = distance_matrix(data[movies], data[movies], p=2)
dd10 = distance_matrix(data[movies], data[movies], p=10) # son datos no compilados en un dataframe.
```

```python
def dm_to_df(dd, col_name):
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(xs = data["star_wars"], ys = data["lord_of_the_rings"], zs=data["harry_potter"])
```

# Clustering jer√°rquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage 
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward")  # Enlace con el cual queremos trabajar.
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90. , leaf_font_size=24.0) #leaf_rotation=90.
plt.show()
```

```python
Z = linkage(data[movies], "average")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], "complete")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
Z
plt.figure(figsize=(25,10))
plt.title("Dendrograma jer√°rquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```
